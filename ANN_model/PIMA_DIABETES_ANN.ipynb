```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/Dell1/Desktop/ML_SUMMER21/Diabetes.csv')
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
mean_ST = data['SkinThickness'].mean(skipna=True)
data = data.replace({'SkinThickness':{0:mean_ST}})
mean_IN = data['Insulin'].mean(skipna=True)
data = data.replace({'Insulin':{0:mean_IN}})
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state=1)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
a = Sequential()
a.add(Dense(16,input_dim=8,activation = 'relu',kernel_initializer='uniform'))
a.add(Dense(10,activation='relu'))
a.add(Dense(6,activation='relu'))
a.add(Dense(4,activation='relu'))
a.add(Dense(1,activation = 'sigmoid'))
#opt = SGD(lr=0.5,momentum=0.5)
a.compile(optimizer= 'adam',loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
a.fit(x_train,y_train,validation_data=(x_test,y_test),epochs = 500)
```

```python
scores = a.evaluate(x_test, y_test, verbose=0)
print("%.2f%%" % ( scores[1]*100))
```

```python
y_pred = a.predict(x_test)
#print(y_pred)
for i in range(len(y_pred)):
    if y_pred[i]>=0.5:
        y_pred[i] = 1
    else:
        y_pred[i] = 0
from sklearn import metrics
cf = metrics.confusion_matrix(y_test,y_pred)
print(cf)
sns.heatmap(cf,annot=True)
```

```python
aaa = metrics.accuracy_score(y_pred,y_test)
print(aaa)
```
